```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
def make_dummies(column_list, dataframe, drop_first=False):
    # Create dummies and replace them with old column
    for column in column_list:
        dummies = pd.get_dummies(dataframe[column], prefix=column, drop_first=drop_first)
        dataframe = pd.concat([dataframe, dummies], axis=1).drop([column], axis=1)
    
    return dataframe
```

```python
# downsample ---check
# create validation set ---check
# Save sets ---check
# Figure out how many trees you need ---check
# Figure out which correlated features to Drop
```

```python
data_raw = pd.read_csv('data/bank-additional-full.csv',delimiter=";")
data_raw = data_raw.drop_duplicates(subset=None, keep='first', inplace=False)
cat_list = ['job', 'marital', 'education', 'default', 'housing', 'loan','contact', 'month', 'day_of_week', 'poutcome']
data = make_dummies(cat_list, data_raw)
data = make_dummies(['y'], data, drop_first=True)
```

```python
# Downsample and create validation set
no_ = data[data['y_yes'] == 0]
no_downsampled = no_.sample(n=4639)
no_val = no_downsampled.sample(frac=0.1)
no_downsampled = no_downsampled.drop(no_val.index)


yes_ = data[data['y_yes'] == 1]
yes_val = yes_.sample(frac=0.1)
yes_ = yes_.drop(yes_val.index)

# training_data , validation_data = pd.concat([no_downsampled ,yes_]) , pd.concat([no_val,yes_val])
training_data = data
```

```python
# Let's see how much trees we need
y = pd.DataFrame(training_data['y_yes'])
x = training_data.copy().drop(['y_yes'], axis=1)
'''s = []
for i in range(1,300, 10):                 
    ensemble_1 = RandomForestClassifier(n_estimators=i, max_features='auto', max_leaf_nodes=None, oob_score=True, n_jobs=-1, random_state=None, verbose=0)
    ensemble_1.fit(x, y)
    s.append(ensemble_1.oob_score_)
s
plt.plot(range(1,300, 10), s)'''
```

```python
# Let's settle for 150
ensemble_1 = RandomForestClassifier(n_estimators=150, max_features='auto', max_leaf_nodes=None, oob_score=True, n_jobs=-1, random_state=None, verbose=0)
ensemble_1.fit(x, y)
pd.Series(ensemble_1.feature_importances_, index=x.columns).nlargest(500).plot(kind='barh', figsize=(10,20))
plt.show()
```

```python
ensemble_1.oob_score_
```

```python
list_of_features = list(pd.Series(ensemble_1.feature_importances_, index=x.columns).nlargest(30).index)
ensemble_2 = RandomForestClassifier(n_estimators=150, max_features='auto', max_leaf_nodes=None, oob_score=True, n_jobs=-1, random_state=None, verbose=0)
ensemble_2.fit(x[list_of_features], y)
pd.Series(ensemble_1.feature_importances_, index=x.columns).nlargest(500).plot(kind='barh', figsize=(10,20))
plt.show()
```
