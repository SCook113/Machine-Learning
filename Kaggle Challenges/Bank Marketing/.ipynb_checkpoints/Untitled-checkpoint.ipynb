```python
def make_dummies(column_list, dataframe, drop_first=False):
    # Create dummies and replace them with old column
    for column in column_list:
        dummies = pd.get_dummies(dataframe[column], prefix=column, drop_first=drop_first)
        dataframe = pd.concat([dataframe, dummies], axis=1).drop([column], axis=1)
    
    return dataframe
```

```python
import pandas as pd
data_raw = pd.read_csv('data/bank-additional-full.csv',delimiter=";")

data_raw = data_raw.drop_duplicates(subset=None, keep='first', inplace=False)

columns_to_do = data_raw.columns
columns_to_do
```

```python
cat_list = ['job', 'marital', 'education', 'default', 'housing', 'loan','contact', 'month', 'day_of_week', 'poutcome']
data = make_dummies(cat_list, data_raw)
pd.set_option('display.max_columns', 200)
data
```
